# kchart Example

```python
import pandas_datareader.data as web
from datetime import datetime
```

```python
import kchart
```

```python
df = web.DataReader('GOOG', 'yahoo', datetime(2017, 1, 1))
```

```python
kchart.draw(df)
```

```python
df = web.DataReader('AAPL', 'yahoo', datetime(2017, 6, 1))
```

```python
kchart.draw(df, 'Apple Inc.')
```

```python
df = web.DataReader('2330.tw', 'yahoo', datetime(2017, 6, 1))
```

```python
df.head()
```

```python
kchart.draw(df, ' 2330')
```

```python
kchart.draw(df, ' 2330', colorup='r', colordown='g')
```
